```python
import glob
import pandas as pd
import numpy as np
```

```python
# 23andMe has 4 versions: v2, v3, v4, and v5
# Each of which has its own build
# In order to match all of them 
# Create a representative list of all 4 datasets
```

```python
g = glob.glob
output_path="snps_list_clean.txt"
```

```python
# Get the 4 files representing each version
files = g("genome_*.txt")
```

```python
# Merge the files
all_data = set()
for f in files:
    data = pd.read_csv(f, sep="\t", comment="#", header=None, names=["rsid", "chr", "pos", "genotype"])
    all_data.update(data["rsid"].astype(str))
```

```python
print(f"Length of merged datasets: {len(all_data)}")
```

```python
# Clean the list
# Remove all enteries with no rsID
# Remove all empty values
all_data = {x for x in all_data if isinstance(x,str) and x.startswith("rs")}
print(f"Clean 23andMe SNP list: {len(all_data)}")
```

```python
with open(output_path, 'w') as out:
    out.write('\n'.join(all_data))
print(f"Saved {len(all_data)} SNPs cleaned and saved")
```
