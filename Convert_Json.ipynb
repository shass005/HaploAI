```python
# Convert txt file into json, relationship is delineated by the number of whitespaces at the begining of the new line
import json

def parse_indented_file(filename):
    with open(filename, 'r') as f:
        lines = [line.rstrip('\n') for line in f if line.strip()]

    stack = []
    root = []

    for line in lines:
        indent = len(line) - len(line.lstrip(' '))
        node = {"name": line.strip(), "children": []}

        while stack and stack[-1][0] >= indent:
            stack.pop()

        if not stack:
            root.append(node)
        else:
            stack[-1][1]["children"].append(node)

        stack.append((indent, node))

    return root

if __name__ == "__main__":
    json_data = parse_indented_file("Phenotype_SNPs.txt")
    with open("Phenotype_SNPs.json", "w") as f:
        json.dump(json_data, f, indent=2)
    print("JSON saved to Phenotype_SNPs.json")
```
